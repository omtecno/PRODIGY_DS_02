```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Downloads/test.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data["Age"] = data["Age"].fillna(data["Age"].mean())
data["Fare"] = data["Fare"].fillna(data["Fare"].median())
data["Cabin"] = data["Cabin"].fillna("Unknown")
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
plt.figure(figsize=(6,5))
sns.histplot(
    data["Age"],
    bins=30,
    kde=False,
    color="teal",
    edgecolor="black"
)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.kdeplot(
    data["Age"],
    fill=True,
    color="purple"
)
plt.title("Age Density Distribution")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(
    x=data["Age"],
    color="orange"
)
plt.title("Age Distribution (Box Plot)")
plt.xlabel("Age")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(data=data, x="Sex")
plt.title("Passenger Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(data=data, x="Pclass")
plt.title("Passenger Distribution by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(
    data=data,
    x="Age",
    y="Fare",
    color="crimson",
    s=60,
    alpha=0.7
)
plt.title("Scatter plot of Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
